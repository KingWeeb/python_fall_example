```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KingWeeb/python_fall_example/main/SolarPrediction.csv', encoding='utf-8')
df = df[['Temperature', 'Humidity', 'Radiation', 'Pressure', 'WindDirection(Degrees)', 'Speed']].astype(dtype=float)
df.head()
```

```python
class DataBot:

    def __init__(self, df):
        self.df = df
        self.run()

    def data_description(self):
        print(f"there are {self.df.shape[0]} rows and {self.df.shape[1]} columns in the dataset")
        print(f"the columns are \n{self.df.columns}")
        print(f"the data types are \n{self.df.dtypes}")
        print(f"the data has {self.df.isnull().sum().sum()} null values")

    def graph(self):
        for col in self.df.columns:
            plt.hist(self.df[col])
            plt.title(col)
            plt.show()
    
    def run(self):
        self.data_description()
        self.graph()
```

```python
DataBot(df)
```
